# Histogram
### Import Libraries and Database

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math

df_test = pd.read_csv('Test_knight.csv')
df_train = pd.read_csv('Train_knight.csv')
```

### Knightâ€™s skills

```python
numeric_columns = df_test.select_dtypes(include='number').columns
num_cols = len(numeric_columns)
cols = 5
rows = math.ceil(num_cols / cols)

sns.set_theme(style="ticks")
palette = sns.color_palette("Paired", n_colors=num_cols)

fig, axes = plt.subplots(rows, cols, figsize=(5*cols, 4*rows))
axes = axes.flatten()

for i, col in enumerate(numeric_columns):
    sns.histplot(
        data=df_test,
        x=col,
        ax=axes[i],
        bins=50,
        color=palette[3],
        label='knight',
        linewidth=0
    )
    axes[i].set_title(col, fontsize=16)
    legend = axes[i].legend()

    axes[i].set_xlabel('')
    axes[i].set_ylabel('')

plt.tight_layout();
```

```python
numeric_columns = df_train.select_dtypes(include='number').columns
num_cols = len(numeric_columns)
cols = 5
rows = math.ceil(num_cols / cols)

sns.set_theme(style="ticks")
palette = sns.color_palette("Paired", n_colors=num_cols)
palette = {palette[5], palette[1]}

fig, axes = plt.subplots(rows, cols, figsize=(5 * cols, 4 * rows))
axes = axes.flatten()

for i, col in enumerate(numeric_columns):
    sns.histplot(
        data=df_train,
        x=col,
        hue='knight',
        ax=axes[i],
        bins=50,
        palette=palette,
        alpha=0.5,
        linewidth=0
    )
    axes[i].set_title(col, fontsize=18)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('')
    
    legend = axes[i].get_legend()
    legend.set_title('')

plt.tight_layout();
```
