```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df_test = pd.read_csv('../../Test_knight.csv')
df_train = pd.read_csv('../../Train_knight.csv')
```

```python
palette = sns.color_palette("Paired", n_colors=10)
palette = {"Jedi": palette[1], "Sith": palette[5]}

fig, axes = plt.subplots(2, 2, figsize=(12, 12))


sns.scatterplot(
    data=df_train,
    x='Empowered',
    y='Stims',
    hue='knight',
    hue_order=['Jedi', 'Sith'],
    palette=palette,
    alpha=0.5,
    linewidth=0,
    ax=axes[0, 0]
)
axes[0, 0].set_xlabel('Empowered')
axes[0, 0].legend(title='')

sns.scatterplot(
    data=df_train,
    x='Empowered',
    y='Stims',
    color='green',
    alpha=0.5,
    linewidth=0,
    ax=axes[1, 0]
)
axes[1, 0].set_xlabel('Empowered')
axes[1, 0].scatter([], [], c='green', alpha=0.5, label='knight')
axes[1, 0].legend(title='')

sns.scatterplot(
    data=df_train,
    x='Friendship',
    y='Survival',
    hue='knight',
    hue_order=['Jedi', 'Sith'],
    palette=palette,
    alpha=0.5,
    linewidth=0,
    ax=axes[0, 1]
)
axes[0, 1].set_xlabel('Friendship')
axes[0, 1].legend(title='')

sns.scatterplot(
    data=df_train,
    x='Friendship',
    y='Survival',
    color='green',
    alpha=0.5,
    linewidth=0,
    ax=axes[1, 1]
)
axes[1, 1].set_xlabel('Friendship')
axes[1, 1].scatter([], [], c='green', alpha=0.5, label='knight')
axes[1, 1].legend(title='')


plt.tight_layout();
```
