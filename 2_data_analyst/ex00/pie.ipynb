# Pie
### Import Libraries and Set Up Database Connection

```python
from sqlalchemy import create_engine, text
import matplotlib.pyplot as plt

db_url = "postgresql+psycopg2://mdesmart:mysecretpassword@localhost:5432/piscineds"
engine = create_engine(db_url)
```

### Extract the data

```python

query = text("SELECT event_type, COUNT(*) FROM customers GROUP BY event_type")

with engine.connect() as connection:
    result = connection.execute(query)
    rows = result.fetchall()

event_types = [row[0] for row in rows]
event_counts = [row[1] for row in rows]
```

### Create the Pie Chart

```python
plt.figure()
plt.pie(event_counts, labels=event_types, autopct='%1.1f%%')
plt.title('Distribution of Event Types')
plt.axis('equal');
```
