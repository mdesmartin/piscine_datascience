# Heatmap
### Import Libraries and Database

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df_test = pd.read_csv('../Test_knight.csv')
df_train = pd.read_csv('../Train_knight.csv')
```

### Corelation matrix

```python
df_train['knight'] = df_train['knight'].map({'Sith': 0, 'Jedi': 1})

correlation_matrix = df_train.corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, cmap='Blues_r', vmin=-.8, vmax=1, center=0.2,
            square=True, linewidths=0, cbar_kws={"shrink": .5})

plt.title('Correlation Matrix Heatmap')
plt.show()
```
